(A) Removing missing records either row-wise or col-wise

```python
df2.dropna()        # by default it removes all those ROWS that contain missing values ; it displays only those rows that have all records in them i.e. do not contain any missing values
# by default axis='rows'
# could write inplace=True to make changes in actual data
```

```python
df2.dropna(axis=1)      #column-wise removal
```

(B) Filling missing values with a constant value

```python
df2.fillna("Pearl")     #fills values instead of null 
```

```python
df2['B'].fillna("Upflairs")     #to fill a particular column
```

NOTE:
1. If col is categorical , then make sure to fill categorical constant or mode value.
2. If col is numerical , then make sure to fill numerical constant or mean or median value.

```python
df2.fillna(df2.mode().iloc[0])      #Mode() returns a dataframe , so we use iloc[0] to select first mode
```

```python
df2
```

```python
df2.interpolate()       #estimate missing values
```

```python
df2['C'].mean()
```

```python
df2['C'].fillna(df2['C'].mean())        #known as DATA IMPUTATION
```

NOTE:
IF a particular col has normal distribution then fill mean() , otherwise in all other distributions fill median().

<h4><b>SOME EXTRA FUNCTIONS</b></h4>

1. CONCAT()

```python
d1={"Name":["Somya","Kush"],"Id":[1,2]}
df1=pd.DataFrame(d1)
d2={"Branch":["CSE","IT"],"Age":[19,20]}
df2=pd.DataFrame(d2)
d=pd.concat([df1,df2])
d
```

2. MAP()

```python
df = pd.DataFrame(
    [('carrot', 'red', 1), 
     ('papaya', 'yellow', 0),
     ('mango', 'yellow', 0), 
     ('apple', 'red', 0)
    ], 
columns=['species', 'color', 'type']
)
mappings = {'carrot': 'veg',
            'papaya': 'fruit'
}
df['type_name'] = df['species'].map(mappings)
df
```

3. APPLY()

```python
def calc_sum(x):        
    return x.sum()

data = {
  "x": [50, 40, 30],
  "y": [300, 500, 200],
  "z": [10,10,10]
}

df = pd.DataFrame(data)
x = df.apply(calc_sum)      #allow the users to pass a function and apply it on every single value
x       
```

4. MERGE()

```python
data1 = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}
data2 = {
  "name": ["Sally", "Peter", "Micky"],
  "age": [77, 44, 22]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df = df1.merge(df2, how='right')        #updates the content of one DataFrame with the content from another DataFrame
df
```

```python
data1 = {
  "name": ["Sally", "Mary", "John"],
  "age": [50, 40, 30]
}
data2 = {
  "name": ["Sally", "Peter", "Micky"],
  "age": [77, 44, 22]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df = df1.merge(df2, how='left')        
df
```
